```python
from sklearn.preprocessing import normalize
import numpy as np
import faiss
import json
# åˆå§‹åŒ–ä¸€ä¸ªç©ºåˆ—è¡¨æ¥å­˜å‚¨æ¯ä¸ªæ–‡æœ¬å—çš„åµŒå…¥å‘é‡
embeddings = []

# éåŽ†æ¯ä¸ªæ–‡æœ¬å—
for chunk in chunks:
    response = get_embedding(chunk)
    if isinstance(response, str):
        response = json.loads(response)
    embedding = np.array(response.data[0].embedding, dtype=np.float32) 
    embeddings.append(embedding)

# ä½¿ç”¨ sklearn çš„ normalize å‡½æ•°å¯¹åµŒå…¥å‘é‡è¿›è¡Œå½’ä¸€åŒ–å¤„ç†
normalized_embeddings = normalize(np.array(embeddings))

# èŽ·å–åµŒå…¥å‘é‡çš„ç»´åº¦
d = len(embeddings[0])

# åˆ›å»ºä¸€ä¸ª Faiss ç´¢å¼•ï¼Œç”¨äºŽå­˜å‚¨å’Œæ£€ç´¢åµŒå…¥å‘é‡
index = faiss.IndexFlatIP(d)

# å°†å½’ä¸€åŒ–åŽçš„åµŒå…¥å‘é‡æ·»åŠ åˆ°ç´¢å¼•ä¸­
index.add(normalized_embeddings)

# èŽ·å–ç´¢å¼•ä¸­çš„å‘é‡æ€»æ•°
n_vectors = index.ntotal


print(n_vectors)
```

```python
from sklearn.preprocessing import normalize
def match_text(input_text, index, chunks, k=2):
    """
    åœ¨ç»™å®šçš„æ–‡æœ¬å—é›†åˆä¸­ï¼Œæ‰¾åˆ°ä¸Žè¾“å…¥æ–‡æœ¬æœ€ç›¸ä¼¼çš„å‰kä¸ªæ–‡æœ¬å—ã€‚

    å‚æ•°:
        input_text (str): è¦åŒ¹é…çš„è¾“å…¥æ–‡æœ¬ã€‚
        index (faiss.Index): ç”¨äºŽæœç´¢çš„Faissç´¢å¼•ã€‚
        chunks (list of str): æ–‡æœ¬å—çš„åˆ—è¡¨ã€‚
        k (int, optional): è¦è¿”å›žçš„æœ€ç›¸ä¼¼æ–‡æœ¬å—çš„æ•°é‡ã€‚é»˜è®¤å€¼ä¸º2ã€‚

    è¿”å›ž:
        str: æ ¼å¼åŒ–çš„å­—ç¬¦ä¸²ï¼ŒåŒ…å«æœ€ç›¸ä¼¼çš„æ–‡æœ¬å—åŠå…¶ç›¸ä¼¼åº¦ã€‚
    """
    # ç¡®ä¿kä¸è¶…è¿‡æ–‡æœ¬å—çš„æ€»æ•°
    k = min(k, len(chunks))

    # ä½¿ç”¨OpenAI APIä¸ºè¾“å…¥æ–‡æœ¬åˆ›å»ºåµŒå…¥å‘é‡
    response = get_embedding(input=input_text)
    if isinstance(response, str):
        response = json.loads(response)
    embedding = np.array(response.data[0].embedding, dtype=np.float32) 
    embeddings.append(embedding)  
    input_embedding = normalize(np.array(embeddings))

    # åœ¨ç´¢å¼•ä¸­æœç´¢ä¸Žè¾“å…¥åµŒå…¥å‘é‡æœ€ç›¸ä¼¼çš„kä¸ªå‘é‡
    distances, indices = index.search(input_embedding, k)
    # åˆå§‹åŒ–ä¸€ä¸ªå­—ç¬¦ä¸²æ¥å­˜å‚¨åŒ¹é…çš„æ–‡æœ¬
    matching_texts = ""
    # éåŽ†æœç´¢ç»“æžœ
    for i, idx in enumerate(indices[0]):
        # æ‰“å°æ¯ä¸ªåŒ¹é…æ–‡æœ¬å—çš„ç›¸ä¼¼åº¦å’Œæ–‡æœ¬å†…å®¹
        print(f"similarity: {distances[0][i]:.4f}\nmatching text: \n{chunks[idx]}\n")
        # å°†ç›¸ä¼¼åº¦å’Œæ–‡æœ¬å†…å®¹æ·»åŠ åˆ°åŒ¹é…æ–‡æœ¬å­—ç¬¦ä¸²ä¸­
        matching_texts += f"similarity: {distances[0][i]:.4f}\nmatching text: \n{chunks[idx]}\n"

    # è¿”å›žåŒ…å«åŒ¹é…æ–‡æœ¬å—åŠå…¶ç›¸ä¼¼åº¦çš„å­—ç¬¦ä¸²
    return matching_texts
```
